# Set up Our World In Data Historical electricity demand

```python
def three_2_two_digits_country(three_code_country):
    """
    Convert 3-digit to 2-digit country code:

    Parameters
    ----------
    three_code_country: str
        3-digit country name

    Returns
    ----------
    two_code_country: str
        2-digit country name
    """
    if three_code_country == "SEN-GMB":
        return f"{three_2_two_digits_country('SN')}-{three_2_two_digits_country('GM')}"

    two_code_country = coco.convert(three_code_country, to="ISO2")
    return two_code_country
```

```python
df_hist = pd.read_csv(r"ourworldindata/electricity-demand.csv")

# interest:
entity = {"Brunei", "Cambodia", "Indonesia", "Laos", "Malaysia", 
          "Myanmar", "Philippines", "Singapore", "Thailand", "Vietnam"}
df_hist = df_hist.query("Entity in @entity")

df_hist["Code"] = df_hist["Code"].map(lambda x: three_2_two_digits_country(x))
```

# Set up IRENA data (1.5c 2050 electricity demand)

```python
df_IRENA = pd.read_excel(r"IRENA-Power-BI/IRENA-Power-BI-Export-Import.xlsx","Power Flow")

df_IRENA = df_IRENA.set_index("Country")
df_IRENA["IRENA_elec_demand"] = df_IRENA["Generation"] + df_IRENA["Net Export (-)/lmport (+)"]
df_IRENA = df_IRENA.drop(columns={"Generation","Net Export (-)/lmport (+)"})

df_IRENA
```

# Combine both Historical data and IRENA

```python
df_int = pd.DataFrame(index=range(2000,2051),columns=df_IRENA.index)

# Put in IRENA data
for country in df_int.columns:
    df_int.loc[2050,country] = df_IRENA.loc[country,"IRENA_elec_demand"]

# Put in Historical data
for country in df_int.columns:
    df_hist_c = df_hist.query("Code == @country")
    for year in df_hist_c.Year.unique():
        df_int.loc[year,country] = df_hist_c.loc[df_hist.Year == year,"Electricity demand - TWh"].item()

# Interpolate
for country in df_int.columns:
    df_int[country] = df_int[country].infer_objects(copy=False).interpolate(method='polynomial', order=2)
```

```python
color_country = {
    "Myanmar": "#440154",
    "Thailand": "#43398e",
    "Malaysia": "#3953a4",
    "Laos": "#1f9d8a",
    "Singapore": "#3fbc73",
    "Cambodia": "#84d44b",
    "Vietnam": "#d8e219",
    "Brunei": "#f3758d",
    "Indonesia": "#c00000",
    "Philippines": "#8a0000",
    }

country_longname = {
    "MM": "Myanmar",
    "TH": "Thailand",
    "MY": "Malaysia",
    "LA": "Laos",
    "SG": "Singapore",
    "KH": "Cambodia",
    "VN": "Vietnam",
    "BN": "Brunei",
    "PH": "Philippines",
    "ID": "Indonesia"
    }

country_prefered_order = ["Myanmar","Thailand","Malaysia","Laos","Singapore","Cambodia","Vietnam","Brunei","Indonesia","Philippines"]

country_prefered_order = ["Philippines","Indonesia","Brunei","Vietnam","Cambodia","Singapore","Laos","Malaysia","Thailand","Myanmar"]

#country_prefered_order = ['Indonesia', 'Singapore', 'Malaysia', 'Brunei', 'Philippines', 'Cambodia',
#                           'Thailand', 'Vietnam', 'Laos', 'Myanmar']

def add_and_rename_countries(df):
    for i in list(country_longname.keys() - set(df.index)):
        df.loc[i] = pd.Series().fillna(0)

    # convert abbreviation into nice long names
    df["Countries"] = [country_longname[c] for c in df.index]
    df = df.set_index("Countries")

    return df
```

```python
df_plot = df_int.copy(deep=True)
df_plot = df_plot.rename(columns=country_longname)

#df_plot = add_and_rename_countries(df_plot)

new_index_order = [c for c in country_prefered_order if c in df_plot.columns]

#new_index_order = [country_longname[c] for c in new_index_order]

fig, ax = plt.subplots(figsize=(5,5))

df_plot.loc[:,new_index_order].plot.area(ax=ax,color=[color_country[c] for c in df_plot.loc[:,new_index_order].columns])

ax.set_xlim([2020, 2050])
ax.set_xlabel("year")
ax.set_ylabel("TWh/year")
ax.set_xticks([2020,2025,2030,2035,2040,2045,2050])
ax.grid()

# Custom legend elements
legend_elements = []

for country in list(color_country.keys()):
    legend_elements = legend_elements + [
        plt.fill_between(
            [],
            [],
            color=color_country[country],
            label=country + ": \n " + str(round(df_plot.T[2050].T[country],2)),
        )
    ]

plt.legend(handles=legend_elements, loc="center left", bbox_to_anchor=(1, 0.5), frameon = False, title = "Energy Demand \n in 2050 [TWh/year]", title_fontproperties={'weight':'bold'})
plt.savefig("load_scale.svg",bbox_inches='tight')
```

# Set up SSP prediction energy curve

```python
scenario_year = {2050, 2040, 2030}
weather_year = 2018

df_ssp = pd.DataFrame(columns=["BN", "KH", "ID", "LA", "MY", "MM", "PH", "SG", "TH", "VN"])

for year in scenario_year:
    df_asia = pd.read_csv(f"../pypsa-earth/data/ssp2-2.6/{year}/era5_{weather_year}/Asia.csv",sep = ";")
    
    # Transplant KH electricity demand for LA
    df_asia.loc[df_asia.region_code == "LA", "Electricity demand"] = df_asia.query("region_code == 'KH'")["Electricity demand"].to_numpy()
    
    df_asia = df_asia.drop(columns=[]).groupby(df_asia.region_code).sum().drop(columns=['region_name','time','region_code'])
    df_asia["Electricity demand"] = df_asia["Electricity demand"] / 1e6 # MWh to TWh
    
    region_code = {"BN", "KH", "ID", "LA", "MY", "MM", "PH", "SG", "TH", "VN"}
    df_asia = df_asia.query("index in @region_code").rename(columns={"Electricity demand":year})

    df_ssp = pd.concat([df_ssp, df_asia.transpose()])

#df_asia.to_csv(f"../pypsa-earth/data/ssp2-2.6/{scenario_year}/era5_{weather_year}/Asia_LA.csv",sep = ";")
df_ssp = df_ssp.sort_index()

df_ssp.loc[2020] = df_ssp.loc[2030]
df_ssp.loc[2025] = df_ssp.loc[2030]
df_ssp.loc[2035] = df_ssp.loc[2040]
df_ssp.loc[2045] = df_ssp.loc[2050]
```
